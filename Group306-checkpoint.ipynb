```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from windrose import WindroseAxes
from io import StringIO
import requests
import gzip

# Define the station IDs and their corresponding URLs
station_ids = ['44062h2016', '44062h2018', 'rplv2h2016','rplv2h2018','fskm2h2016','fskm2h2018']
urls = [
    'https://www.ndbc.noaa.gov/data/historical/stdmet/44062h2016.txt.gz',
    'https://www.ndbc.noaa.gov/data/historical/stdmet/44062h2018.txt.gz',
    'https://www.ndbc.noaa.gov/data/historical/stdmet/rplv2h2016.txt.gz', 'https://www.ndbc.noaa.gov/data/historical/stdmet/rplv2h2018.txt.gz',
    'https://www.ndbc.noaa.gov/data/historical/stdmet/fskm2h2016.txt.gz','https://www.ndbc.noaa.gov/data/historical/stdmet/fskm2h2018.txt.gz'
]

# Function to create wind roses using the windrose package
def create_wind_rose(df, year, station_id):
    # Determine appropriate wind speed bins based on the data
    max_wind_speed = df['WSPD'].max()
    bins = np.linspace(0, max_wind_speed, 6)  # Create 6 bins from 0 to max_wind_speed
    labels = [f'{int(bins[i])}-{int(bins[i+1])} m/s' for i in range(len(bins)-1)]
    
    # Create the wind rose plot
    ax = WindroseAxes.from_ax()
    ax.bar(df['WDIR'], df['WSPD'], normed=True, opening=0.8, edgecolor='white', bins=bins)
    
    # Configure the plot
    ax.set_title(f'Wind Rose for {station_id} in {year}')
    ax.set_legend(title='Wind Speed (m/s)', labels=labels)
    plt.show()

# Function to convert days since New Year to a proper date
def convert_to_date(year, days):
    return pd.Timestamp(year, 1, 1) + pd.Timedelta(days=int(days) - 1)

# Iterate through each URL and fetch the data
for station_id, url in zip(station_ids, urls):
    # Send a GET request to the NDBC server
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Decompress the gzip file
        data = gzip.decompress(response.content).decode('utf-8')
        
        # Load the data into a pandas DataFrame
        df = pd.read_csv(StringIO(data), delim_whitespace=True, skiprows=[1])
        
        # Filter out rows where WDIR or WSPD are 999 or WSPD is 99 or greater
        df = df[(df['WDIR'] != 999) & (df['WSPD'] != 999) & (df['WSPD'] < 99)]
        
        # Convert 'days since New Year' to proper dates
        df['Date'] = df.apply(lambda row: convert_to_date(int(row['#YY']), row['DD']), axis=1)
        
        # Extract the year
        df['Year'] = df['Date'].dt.year
        
        # Group by year and create wind roses
        for year, group in df.groupby('Year'):
            create_wind_rose(group, year, station_id)
        
        # Save the filtered DataFrame to a CSV file
        df.to_csv(f'{station_id}_filtered_data.csv', index=False)
    else:
        print(f'Failed to retrieve data for station {station_id}. HTTP Status code: {response.status_code}')
```

```python

import requests
import pandas as pd
from io import StringIO
import gzip
import matplotlib.pyplot as plt
import numpy as np

# Define the station IDs and their corresponding URLs
station_ids = ['44062h2016', '44062h2018']
urls = [
    'https://www.ndbc.noaa.gov/data/historical/stdmet/44062h2016.txt.gz',
    'https://www.ndbc.noaa.gov/data/historical/stdmet/44062h2018.txt.gz'
]

# Function to create wind roses using Matplotlib
def create_wind_rose(df, year, station_id):
    # Define the number of bins and the angle steps
    num_bins = 8
    angle_steps = np.arange(0, 360, 360 / num_bins)

    # Define the wind speed bins
    speed_bins = [0, 1, 3, 6, 10, 20]

    # Calculate histogram
    hist, xedges, yedges = np.histogram2d(df['WDIR'], df['WSPD'], bins=[angle_steps, speed_bins])

    # Create the wind rose plot
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, polar=True)

    # Plot each bin
    for i in range(len(speed_bins) - 1):
        ax.bar(angle_steps[:-1] * np.pi / 180, hist[:, i], width=(2 * np.pi / num_bins), bottom=hist[:, :i].sum(axis=1))

    # Configure the plot
    ax.set_theta_zero_location('N')
    ax.set_theta_direction(-1)
    ax.set_title(f'Wind Rose for {station_id} in {year}')
    plt.legend(['0-1 m/s', '1-3 m/s', '3-6 m/s', '6-10 m/s', '10-20 m/s'], loc='upper right')
    plt.show()

# Function to convert days since New Year to a proper date
def convert_to_date(year, days):
    return pd.Timestamp(year, 1, 1) + pd.Timedelta(days=int(days) - 1)

# Iterate through each URL and fetch the data
for station_id, url in zip(station_ids, urls):
    # Send a GET request to the NDBC server
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Decompress the gzip file
        data = gzip.decompress(response.content).decode('utf-8')
        
        # Load the data into a pandas DataFrame
        df = pd.read_csv(StringIO(data), delim_whitespace=True, skiprows=[1])
        
        # Filter out rows where WDIR or WSPD are 999
        df = df[(df['WDIR'] != 999) & (df['WSPD'] != 999)]
        
        # Convert 'days since New Year' to proper dates
        df['Date'] = df.apply(lambda row: convert_to_date(int(row['#YY']), row['DD']), axis=1)
        
        # Extract the year
        df['Year'] = df['Date'].dt.year
        
        # Group by year and create wind roses
        for year, group in df.groupby('Year'):
            create_wind_rose(group, year, station_id)
        
        # Save the filtered DataFrame to a CSV file
        df.to_csv(f'{station_id}_filtered_data.csv', index=False)
    else:
        print(f'Failed to retrieve data for station {station_id}. HTTP Status code: {response.status_code}')
```
